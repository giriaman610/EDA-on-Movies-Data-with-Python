```python
#plotting histogram for each column

plt.figure(figsize=(6,4))
for col in numeric_columns:
    sns.histplot(data=df, x=col,bins=50, kde=True)
    plt.title(f'{col}_distribution')
    plt.show()
```

```python
# checking some groupby aggregations
genre_year_stats=df.groupby(['Genre','ReleaseYear']).agg({
    "Global_BoxOfficeUSD" :'mean',
    "BudgetUSD" : 'mean',
    'IMDbRating' : 'mean',
    'RottenTomatoesScore' : 'mean'
}
).reset_index()
```

```python
genre_year_stats
```

```python
genre_year_stats=df.groupby(['Genre']).agg({
    "Global_BoxOfficeUSD" :'mean',
    "BudgetUSD" : 'mean',
    'IMDbRating' : 'mean',
    'RottenTomatoesScore' : 'mean'
}
).reset_index()
```

```python
genre_year_stats['Budget_Millions'] = (genre_year_stats['BudgetUSD'] / 1e6).round(2)
genre_year_stats['Global_BoxOfficeUSD_Millions'] = (genre_year_stats['Global_BoxOfficeUSD'] / 1e6).round(2)
genre_year_stats
```

```python
plt.figure(figsize=(14,8))
sns.lineplot(data=genre_year_stats, x='ReleaseYear', y='Global_BoxOfficeUSD', hue='Genre')
plt.yscale('log')
plt.title("Avg  Global box Office by Genre over years")
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.lineplot(data=genre_year_stats.query("Genre== 'Horror'"), x='ReleaseYear', y='Global_BoxOfficeUSD', hue='Genre')
plt.yscale('log')
plt.title("Avg  Global box Office by Genre over years")
plt.show()
```
